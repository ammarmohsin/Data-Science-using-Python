## Checking Row wise data

### Using Method "loc and iloc"

- loc gets rows (or columns) with particular labels from the index

- iloc gets rows (or columns) at particular postions in the index (so it only takes integers)

```python
# Accessing the first row
sales.loc[0]
```

```python
# 100th row

sales.loc[99]
```

```python
# this will fail in 'loc' method but it will work in 'iloc' method
sales.iloc[-1]
```

```python
# iloc Method

sales.iloc[[-1, -3]] # Multiple row data
```

```python
sales.tail()
```

### Particular Rows and Columns

```python
# using loc method
sales.loc[[0,99,999], ['Country', 'Age_Group', 'Product_Category']] # it will work only for the column names not index
```

```python
# using iloc method

sales.iloc[[9,99,999], [7,5,9]] # it will work only for index not the column names
```

```python
# Get the under 25 customers sales data of the United States

# Method-1
sales.loc[(sales['Country'] == 'United States') & (sales['Age_Group'] == 'Youth (<25)')]
```

```python
# Method-2

# ":" in python means give me all the columns

sales.loc[(sales['Country'] == 'United States') & (sales['Age_Group'] == 'Youth (<25)'), :]
```

```python
profit_mean = sales['Profit'].mean()
profit_mean
```

```python
# Select all data entries against which profit is greater than mean (average) value

# Method-1
sales.loc[sales['Profit'] > profit_mean]
```

## Grouped and Aggregated Calculations

### Use of "groupby" function

```python
# Group the data country wise
sales.groupby("Country")
```

```python
# How many time each particular country value repeated in data?
sales.groupby("Country").count()
```

```python
# Computing average profit country wise
sales.groupby('Country')['Profit'].mean()
```

```python
# Computing average profit Age_Group wise

sales.groupby('Age_Group')['Profit'].mean()
```

```python
# Computing average of Profit and ordered quantity in particular month
sales.groupby('Month')[["Profit", "Order_Quantity"]].mean()
```

```python
# Computing median of Profit and ordered quantity in particular month
sales.groupby('Month')[["Profit", "Order_Quantity"]].median()
```

```python
# Computing average of Profit and Ordered Quantity in Particular month and Age group wise
sales.groupby(['Month', 'Age_Group'])[["Profit", "Order_Quantity"]].mean()
```

```python
# Write in Multiple lines for better interpetation
(sales.
groupby(['Month', 'Age_Group'])[["Profit", "Order_Quantity"]].
mean())
```

```python
# Find number of unqiue sub_categories in each Product_Category
sales.groupby('Product_Category')['Sub_Category'].nunique()
```

```python
p_c = sales.groupby('Product_Category')['Profit'].mean()
p_c
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Country wise Bar Chart w.r.t average and std deviation of the Order Quantity

sal = sales.groupby('Country')
mean = sales[['Country', 'Order_Quantity']].groupby('Country').mean()
error = sales[['Country', 'Order_Quantity']].groupby('Country').std()

fig, sal = plt.subplots()
mean.plot.bar(yerr= error, ax= sal, capsize= 4, figsize= (15,8))
sal.set_ylabel("Average Product Order Quantity")
sal.set_xlabel("Country")
sal.set_title("Country wise Average Order Quantity with Standard Deviation bar")
```
